Поиск 5 (можно настроить любое число в параметре max_files) файлов с максимальным количеством ненулевых значений diff в values из архива

```python
import os
import pandas as pd
import json
import glob
import zipfile
from heapq import nlargest

folder_path = '/content/drive/MyDrive/json_fs17/fs17.zip'
max_files = 20  # Количество файлов с наибольшими значениями

# Извлечение файлов из архива
with zipfile.ZipFile(folder_path, 'r') as zip_ref:
    zip_ref.extractall('/content/tmp')

# Получить список JSON-файлов в извлеченной папке
extracted_folder = '/content/tmp'
json_files = glob.glob(os.path.join(extracted_folder, '*.json'))

# Словарь для хранения соответствия между именем файла и датафреймом
file_dataframes = {}

# Цикл по файлам
for file in json_files:
    # Проверить размер файла
    file_size = os.path.getsize(file)
    if file_size < 100 * 1024:
        continue

    # Прочитать содержимое файла
    with open(file, 'r') as f:
        json_data = f.read()

    # Распарсить JSON и создать датафрейм
    data = json.loads(json_data)
    result_values = data['data']['result'][0]['values']
    df = pd.DataFrame(result_values, columns=['timestamp', 'value'])

    # Преобразовать значения в столбце 'value' в числовой формат
    df['value'] = df['value'].astype(float)

    # Продифференцировать колонку 'value'
    df['diff'] = df['value'].diff()

    # Подсчитать количество ненулевых элементов в столбце 'diff'
    nonzero_count = df['diff'].astype(bool).sum()

    # Добавить в словарь соответствие между именем файла и датафреймом
    file_dataframes[file] = df
"""
    # Вывести датафрейм и количество ненулевых элементов
    print(f"Файл: {file}")
    print(df)
    print("Количество ненулевых элементов в столбце 'diff':", nonzero_count)
    print('-' * 50)
"""
# Получить пять датафреймов с наибольшими значениями nonzero_count
top_files = nlargest(max_files, file_dataframes, key=lambda x: file_dataframes[x]['diff'].astype(bool).sum())

# Вывести только пять датафреймов с наибольшими значениями nonzero_count
print("Пять датафреймов с наибольшими значениями nonzero_count:")
for file in top_files:
    print(f"Файл: {file}")
    print(file_dataframes[file])
    print('-' * 50)

# Удалить извлеченные файлы
for file in json_files:
    os.remove(file)
os
```
